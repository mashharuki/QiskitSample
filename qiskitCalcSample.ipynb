```python
from qiskit import IBMQ

# アカウント情報を読み込む
IBMQ.load_account()
# 設定を適用する
provider = IBMQ.get_provider(group='open', project='main')
# マシンを選択する。
backend = provider.get_backend('ibmq_5_yorktown')
backend.configuration().basis_gates
```

```python
import numpy as np
from qiskit import QuantumCircuit

# 量子ビットを用意する。
qc = QuantumCircuit(3, 1)
# ゲートを配置する。
qc.x(0)
qc.y(1)
qc.z(2)
# 量子ゲートを描画する。
qc.draw('mpl')
```

```python
import qiskit

qc = qiskit.compiler.transpile(qc, basis_gates=['id', 'rz', 'sx', 'x', 'cx'])
# 量子ゲートを描画する。
qc.draw('mpl')
```

```python
qc = qiskit.compiler.transpile(qc, backend=backend)
# 量子ゲートを描画する。
qc.draw('mpl')
```

```python
# 量子ビットを用意する。
qc = QuantumCircuit(3, 1)
# 量子ゲートを設定する。
qc.x(0)
qc.x(1)
qc.rz(np.pi, 1)
qc.rz(np.pi, 2)
# 量子ゲートを描画する
qc.draw('mpl')
```

```python
# 量子ビットを用意する。
qc = QuantumCircuit(3, 1)
# ゲートを配置する。
qc.x(0)
qc.y(1)
qc.z(2)

qc = qiskit.compiler.transpile(qc, basis_gates=['id', 'u1', 'u2', 'u3', 'cx'])
# 量子ゲートを描画する。
qc.draw('mpl')
```
