```python
# 2つの量子ビットによる計算
import numpy as np
from qiskit import QuantumCircuit

# 量子ビットを用意する。
qc = QuantumCircuit(2, 1)
# 量子ゲートを設定する。
qc.cx(0, 1)
qc.barrier()
qc.cy(0, 1)
qc.barrier()
qc.cz(0, 1)
# 測定ゲートを配置する。
qc.measure(0, 0)
qc.draw(output='mpl')
```

```python
# 量子ゲートを分解して描画する。
qc_basis2 = qc.decompose()
qc_basis2.draw(output='mpl')
```

```python
# さらにもう一段階量子ゲートを分解して描画する。
qc_basis3 = qc_basis2.decompose()
qc_basis3.draw(output='mpl')
```

```python
# swapゲートを分解するコード
qc = QuantumCircuit(2)
# swapゲートを配置する。
qc.swap(0, 1)
qc.draw(output='mpl')
```

```python
qc_basis4 = qc.decompose()
qc_basis4 .draw(output='mpl')
```

```python
# トフォリゲートを分解するコード
qc = QuantumCircuit(3)
# ccxゲートを配置する
qc.ccx(0, 1, 2)
qc.draw(output='mpl')
```

```python
# 2段階分解して描画する
qc_basis5 = qc.decompose().decompose()
qc_basis5 .draw(output='mpl')
```
